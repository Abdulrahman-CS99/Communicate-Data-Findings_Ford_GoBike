## Investigation Overview

> In this Investigation, I will look at the characteristics for the duration of trips, the main focus was on age, gender and type of user.
## Dataset Overview

> The dataset contains information about trips in the bike system in February 2019, the dataset has information like user gender and duration of trip etc.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import matplotlib.colors as colors

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('201902-fordgobike-tripdata_clean.csv')
```

## (The duration of trip by user type)

> The chart is to show you which type of users finsih the trip more faster.And it looks likle subscribers are faster.

```python
base_color=sb.color_palette()[0]
sb.boxplot(data=df,x='user_type',y='duration_min',color=base_color)
plt.xticks(rotation=15)
plt.ylim(0,40);
plt.xlabel('user type')
plt.ylabel('duration in minutes');
plt.title('The duration of trip by user type');
```

## (The duration of trip by age)

> The chart shows which age is most frequent in using our bikes also it illustrates which age spend more time in trips. It looks like younger ages spend more time and the most frequent range of age is between 20-40.


```python
plt.figure(figsize=[8,6])

plt.legend(['Subcriber','Customer','S','A','W','Q','X']);

bins_x=np.arange(18,100,10)
bins_y=np.arange(1,1500,250)
plt.hist2d(data=df,x='age',y='duration_min',bins=[bins_x,bins_y])
plt.colorbar(label='More brighter colour indicates more data points in the cell');
plt.ylim(0.1500)
plt.xlabel('age')
plt.ylabel('duration in minutes');
plt.title('The duration of trip by age');
```

## (The duration of the trip by age in both subscribers and customers type)

> The next chart shows the most age spending time in trips for both user types. It seems like there is no difference in types, all younger ages users spending more duration in the trips for all types.


```python
plt.figure(figsize=[10,6])

userType_marker=[['Subscriber','o'],['Customer','^']]

for userType,marker in userType_marker:
    plot_data=df.loc[df['user_type']==userType]
    sb.regplot(data=plot_data,x='age',y='duration_min',marker=marker,fit_reg=False,scatter_kws={"alpha" : 0.3,"s":18})
plt.xlabel('age')
plt.ylabel('duration in minutes')
plt.legend(['Subcriber','Customer']);
plt.title('The duration of the trip by age in both subscribers and customers type');
```
